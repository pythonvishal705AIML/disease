```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Genetic Algorithm Parameters
POPULATION_SIZE = 20
GENE_LENGTH = 30  # Number of features in your dataset
GENERATIONS = 50
MUTATION_RATE = 0.1

# Fitness Function: Accuracy of the model with selected features
def fitness_function(chromosome, X_train1, X_test1, y_train1, y_test1):
    selected_features = [i for i in range(len(chromosome)) if chromosome[i] == 1]
    if len(selected_features) == 0:  # Avoid empty feature subsets
        return 0
    X_train_sel = X_train1[:, selected_features]
    X_test_sel = X_test1[:, selected_features]
    model = RandomForestClassifier()
    model.fit(X_train_sel, y_train1)
    predictions = model.predict(X_test_sel)
    return accuracy_score(y_test1, predictions)

# Generate Initial Population
def initialize_population():
    return np.random.randint(2, size=(POPULATION_SIZE, GENE_LENGTH))

# Selection (Roulette Wheel Selection)
def select_parents(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    probabilities = [fitness / total_fitness for fitness in fitness_scores]
    parents = population[np.random.choice(range(POPULATION_SIZE), size=2, p=probabilities)]
    return parents

# Crossover (Single-Point Crossover)
def crossover(parent1, parent2):
    point = np.random.randint(1, GENE_LENGTH)
    child1 = np.concatenate((parent1[:point], parent2[point:]))
    child2 = np.concatenate((parent2[:point], parent1[point:]))
    return child1, child2

# Mutation
def mutate(chromosome):
    for i in range(len(chromosome)):
        if np.random.rand() < MUTATION_RATE:
            chromosome[i] = 1 - chromosome[i]  # Flip the bit
    return chromosome

# Main Genetic Algorithm Loop
def genetic_algorithm(X, y):
    population = initialize_population()
    X_train1, X_test1, y_train1, y_test1 = train_test_split(X, Y, test_size=0.3, random_state=42)
    best_chromosome = None
    best_fitness = 0

    for generation in range(GENERATIONS):
        fitness_scores = [fitness_function(chromosome, X_train1, X_test1, y_train1, y_test1) for chromosome in population]
        new_population = []
        for _ in range(POPULATION_SIZE // 2):
            parent1, parent2 = select_parents(population, fitness_scores)
            child1, child2 = crossover(parent1, parent2)
            new_population.append(mutate(child1))
            new_population.append(mutate(child2))
        population = np.array(new_population)

        # Track the best chromosome
        max_fitness = max(fitness_scores)
        if max_fitness > best_fitness:
            best_fitness = max_fitness
            best_chromosome = population[np.argmax(fitness_scores)]

        print(f"Generation {generation + 1}, Best Fitness: {best_fitness}")
    # plt.plot(best_fitness,generation + 1)
    # plt.xlabel("Generation")
    # plt.ylabel("Best Fitness (Accuracy)")
    # plt.title("GA Fitness Progression")
    # plt.show()

    return best_chromosome, best_fitness

# Example Dataset (Replace with your own data)
from sklearn.datasets import make_classification
X, Y = make_classification(n_samples=100, n_features=30, n_informative=15, random_state=42)

# Run Genetic Algorithm
best_chromosome, best_fitness = genetic_algorithm(X, Y)
print("Best Chromosome:", best_chromosome)
print("Best Fitness (Accuracy):", best_fitness)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

num_particles = 30
num_iterations = 50
inertia_weight = 0.5  # w
cognitive_constant = 2  # c1
social_constant = 2  # c2

def fitness_function(particle):
    selected_features = X[:, particle.astype(bool)]
    if selected_features.shape[1] == 0:
        return 0

    X_train1, X_test1, y_train1, y_test1 = train_test_split(
        X, Y, test_size=0.3, random_state=42)

    model = RandomForestClassifier()
    model.fit(X_train1, y_train1)
    predictions = model.predict(X_test1)
    return accuracy_score(y_test1, predictions)

def initialize_particles(num_particles, num_features):
    particles = np.random.randint(2, size=(num_particles, num_features))
    velocities = np.random.uniform(-1, 1, (num_particles, num_features))
    return particles, velocities

def particle_swarm_optimization():
    num_features = X.shape[1]
    particles, velocities = initialize_particles(num_particles, num_features)
    personal_best_positions = particles.copy()
    personal_best_scores = np.array([fitness_function(p) for p in particles])
    global_best_position = personal_best_positions[np.argmax(personal_best_scores)]
    global_best_score = np.max(personal_best_scores)

    for iteration in range(num_iterations):
        for i in range(num_particles):

            inertia = inertia_weight * velocities[i]
            cognitive = cognitive_constant * np.random.rand() * (personal_best_positions[i] - particles[i])
            social = social_constant * np.random.rand() * (global_best_position - particles[i])
            velocities[i] = inertia + cognitive + social

            probabilities = 1 / (1 + np.exp(-velocities[i]))
            particles[i] = np.random.rand(num_features) < probabilities

            fitness = fitness_function(particles[i])
            if fitness > personal_best_scores[i]:
                personal_best_scores[i] = fitness
                personal_best_positions[i] = particles[i]
                if fitness > global_best_score:
                    global_best_score = fitness
                    global_best_position = particles[i]

        print(f"Iteration {iteration + 1}: Best Fitness = {global_best_score:.4f}")

    print("\nBest Solution Found:", global_best_position)
    print("Best Fitness:", global_best_score)

particle_swarm_optimization()
```

```python
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

class Agent:
    def __init__(self, id, initial_position):
        self.id = id
        self.position = initial_position  # Represents the agent's choice (e.g., feature subset)
        self.fitness = 0  # Represents the accuracy with this subset

    def evaluate(self, X_train1, X_test1, y_train1, y_test1):
        # Subset the data based on the agent's position
        selected_features = [i for i, val in enumerate(self.position) if val == 1]
        if not selected_features:  # Handle case with no features selected
            self.fitness = 0
            return

        X_train_subset = X_train1[:, selected_features]
        X_test_subset = X_test1[:, selected_features]

        # Train and evaluate a classifier (Random Forest)
        model = RandomForestClassifier(random_state=42, n_estimators=10)  # Or any other model
        model.fit(X_train_subset, y_train1)
        predictions = model.predict(X_test_subset)
        self.fitness = accuracy_score(y_test1, predictions)

    def __repr__(self):
        return f"Agent {self.id}: Position={self.position}, Fitness={self.fitness}"

def upper_split(X, Y, num_agents=20):
    X_train1, X_test1, y_train1, y_test1 = train_test_split(X, Y, test_size=0.3, random_state=42)
    num_features = X.shape[1]
    agents = []

    for i in range(num_agents):
        # Initialize agents with random feature subsets (binary representation)
        initial_position = [random.randint(0, 1) for _ in range(num_features)]
        agents.append(Agent(i, initial_position))

    # Evaluate the fitness of all agents
    for agent in agents:
        agent.evaluate(X_train1, X_test1, y_train1, y_test1)

    # Sort agents based on their fitness in descending order
    agents.sort(key=lambda agent: agent.fitness, reverse=True)

    # Return the best agent
    return agents[0]

# Example usage:
best_agent = upper_split(X, Y)
print(f"\nOverall Best Agent: {best_agent}")
```

```python
import random
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

class Agent:
    def __init__(self, id, initial_position):
        """
        Represents an agent (solution), where 'position' is a binary list
        indicating which features are selected (1) or not (0).
        """
        self.id = id
        self.position = initial_position
        self.fitness = 0           # Classification accuracy on the validation set

        def evaluate(self, X_train, X_valid, y_train, y_valid):
        """
        Subset the data based on the agent's chosen features, then
        train/evaluate a LogisticRegression model on the train/validation sets.
        """
        selected_features = [i for i, val in enumerate(self.position) if val == 1]
        if not selected_features:  # Handle the case with no selected features
            self.fitness = 0
            return

        # Subset the features
        X_train_subset = X_train[:, selected_features]
        X_valid_subset = X_valid[:, selected_features]

        # Train and evaluate a LogisticRegression
        model = LogisticRegression(
            random_state=42,
            max_iter=500,        # Allow enough iterations to converge
            penalty='l2',        # Default L2 regularization
            solver='lbfgs'       # Common solver for small/medium datasets
        )
        model.fit(X_train_subset, y_train)
        predictions = model.predict(X_valid_subset)
        self.fitness = accuracy_score(y_valid, predictions)

    def __repr__(self):
        return f"Agent {self.id}: Position={self.position}, Fitness={self.fitness:.4f}"

def upper_split(X, Y, num_agents=20, generations=50):
    """
    Splits the dataset into Train, Validation, and Test sets.
    Evolves agents to select the best feature subsets according
    to validation-set performance. Returns the best agent.
    """
    # First split into (Train+Validation) and Test
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, Y, test_size=0.3, random_state=42
    )

    # Further split Train+Validation into Train and Validation
    X_train, X_valid, y_train, y_valid = train_test_split(
        X_train_val, y_train_val, test_size=0.3, random_state=42
    )

    num_features = X.shape[1]

    # Initialize agents with random feature subsets
    agents = []
    for i in range(num_agents):
        initial_position = [random.randint(0, 1) for _ in range(num_features)]
        agents.append(Agent(i, initial_position))

    for generation in range(generations):
        # Evaluate each agent on the current Train/Validation sets
        for agent in agents:
            agent.evaluate(X_train, X_valid, y_train, y_valid)

        # Sort agents by fitness (descending order)
        agents.sort(key=lambda agent: agent.fitness, reverse=True)

        print(f"Generation {generation + 1}, Best Agent: {agents[0]}")

        # Elitism: Keep the best agent for the next generation
        new_agents = [agents[0]]

        # Crossover and mutation to generate offspring
        while len(new_agents) < num_agents:
            # Select parents from the top-performing half
            parent1 = random.choice(agents[: num_agents // 2])
            parent2 = random.choice(agents[: num_agents // 2])

            child_position = []
            for i in range(num_features):
                # Crossover
                if random.random() < 0.5:
                    child_position.append(parent1.position[i])
                else:
                    child_position.append(parent2.position[i])

            # Mutation (flip bits with 10% probability)
            for i in range(num_features):
                if random.random() < 0.1:
                    child_position[i] = 1 - child_position[i]

            new_agents.append(Agent(len(new_agents), child_position))

        # Replace old generation with new
        agents = new_agents

    # After the final generation, the best agent is:
    best_agent = agents[0]

    # Evaluate the best agent on the unseen Test set
    selected_features = [i for i, val in enumerate(best_agent.position) if val == 1]
    if not selected_features:
        print("\nNo features selected by best agent; accuracy on Test set = 0")
        return best_agent

    X_test_subset = X_test[:, selected_features]
    # Retrain on the combined Train+Validation to maximize data usage
    final_model = LogisticRegression(
        random_state=42,
        max_iter=500,
        penalty='l2',
        solver='lbfgs'
    )
    X_train_val_subset = X_train_val[:, selected_features]
    test_predictions = final_model.predict(X_test_subset)
    test_fitness = accuracy_score(y_test, test_predictions)

    print(f"\nBest Agent (after {generations} generations): {best_agent}")
    print(f"Test Accuracy with selected features: {test_fitness:.4f}")

    return best_agent


# Example usage (assuming X and Y are defined):
if __name__ == "__main__":
    from sklearn.datasets import load_iris

    data = load_iris()
    X, Y = data.data, data.target

    best_agent = upper_split(X, Y, num_agents=10, generations=20)
```
